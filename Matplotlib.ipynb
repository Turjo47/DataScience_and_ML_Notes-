```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(0, 10*np.pi, 1000)
Y = np.sin(X)
```

```python
fig, ax = plt.subplots()
ax.plot(X,Y)
plt.show()
```

### Different Types of Graphs

```python
#scatter graph
X = np.random.uniform(0,1,100)
Y = np.random.uniform(0,1,100)
plt.scatter(X,Y)
plt.show()
```

```python
#bar chart
X = np.arange(10)
Y = np.random.uniform(0,1,10)
plt.bar(X,Y)
plt.show()
```

```python
#Image data
Z = np.random.uniform(0,1,(8,8))
plt.imshow(Z)
plt.show()
```

```python
#Contour
Z=np.random.uniform(0,1,(8,8))
plt.contourf(Z)
plt.show()
```

```python
#Pie Chart
Z=np.random.uniform(0,1,4)
plt.pie(Z)
plt.show()
```

```python
#Histogram
Z = np.random.normal(0,1,100)
plt.hist(Z)
plt.show()
```

```python
#ErrorBar
X= np.arange(5)
Y= np.random.uniform(0,1,5)
plt.errorbar(X,Y)
plt.show()
```

```python
#Box ploting
Z=np.random.normal(0,1,(100,3))
plt.boxplot(Z)
plt.show()
```

### Tweak
#### can modify anything in plot, including limits,colors,markers,line width and style, ticks labels, titles, etc

```python
#color change
X= np.linspace(0,10,100)
Y= np.sin(X)
plt.plot(X,Y, color="black")
plt.show()
```

```python
#Line style change
X=np.linspace(0,1,100)
Y=np.sin(X)
plt.plot(X,Y, color="red", linestyle="--")
plt.show()
```

```python
#Line thickness
X=np.linspace(0,1,100)
Y=np.sin(X)
plt.plot(X,Y, color="red", linestyle="--", linewidth=5)
plt.show()
```

```python
#Putting Marker
X=np.linspace(0,1,100)
Y=np.sin(X)
plt.plot(X,Y, color="red", linewidth=1, marker="o")
plt.show()
```

### Organize
#### can plot several data on same figure but can also split a figure in several subplots(named Axes)

```python
#Multiple line in same figure
X=np.linspace(0,10,100)
Y1,Y2 = np.sin(X),np.cos(X)
plt.plot(X,Y1,X,Y2)
plt.show()
```
