```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(0, 10*np.pi, 1000)
Y = np.sin(X)
```

```python
fig, ax = plt.subplots()
ax.plot(X,Y)
plt.show()
```

### Different Types of Graphs

```python
#scatter graph
X = np.random.uniform(0,1,100)
Y = np.random.uniform(0,1,100)
plt.scatter(X,Y)
plt.show()
```

```python
#bar chart
X = np.arange(10)
Y = np.random.uniform(0,1,10)
plt.bar(X,Y)
plt.show()
```

```python
#Image data
Z = np.random.uniform(0,1,(8,8))
plt.imshow(Z)
plt.show()
```

```python
#Contour
Z=np.random.uniform(0,1,(8,8))
plt.contourf(Z)
plt.show()
```

```python
#Pie Chart
Z=np.random.uniform(0,1,4)
plt.pie(Z)
plt.show()
```

```python
#Histogram
Z = np.random.normal(0,1,100)
plt.hist(Z)
plt.show()
```

```python
#ErrorBar
X= np.arange(5)
Y= np.random.uniform(0,1,5)
plt.errorbar(X,Y)
plt.show()
```

```python
#Box ploting
Z=np.random.normal(0,1,(100,3))
plt.boxplot(Z)
plt.show()
```
