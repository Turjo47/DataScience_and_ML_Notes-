```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(0, 10*np.pi, 1000)
Y = np.sin(X)
```

```python
fig, ax = plt.subplots()
ax.plot(X,Y)
plt.show()
```

### Different Types of Graphs

```python
#scatter graph
X = np.random.uniform(0,1,100)
Y = np.random.uniform(0,1,100)
plt.scatter(X,Y)
plt.show()
```

```python
#bar chart
X = np.arange(10)
Y = np.random.uniform(0,1,10)
plt.bar(X,Y)
plt.show()
```
